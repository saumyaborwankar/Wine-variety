```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
train=pd.read_csv('D:/My stuff/ML TASK INTERNSHALA/Knight ML Assignment/Data/train.csv')
test=pd.read_csv('D:/My stuff/ML TASK INTERNSHALA/Knight ML Assignment/Data/test.csv')
```

```python
##### bar plot 
train=train.drop(['user_name','region_2'],axis=1)
train['country'].value_counts().head(10).plot.bar()
```

```python
import matplotlib as mlt

ax = train['points'].value_counts().sort_index().plot.bar(
    figsize = (12,6),
    fontsize = 14
)
ax.set_title("Count the Wine-Points", fontsize = 20)
ax.set_xlabel("Points", fontsize = 20)
ax.set_ylabel("Count", fontsize = 20)
```

```python
#plot price
train['price'].value_counts().sort_index().plot.line()
```

```python
train['points'].value_counts().sort_index().plot.area()
```

```python
#Scatter Plot
train.plot.scatter(x='points', y='price')
train.isnull().count()
```

```python
import seaborn as sns
sns.countplot(train['province'].head(20))
```

```python
top_5 = train.groupby('country').mean().sort_values(by = ['points'],ascending = False)
top_5 = pd.DataFrame(top_5.iloc[0:5,0])
top_5['Index'] = list(range(5))
top_5['Country'] = top_5.index.values
top_5.set_index(['Index'],inplace = True)
top_5
```

```python
top_5 = train.groupby('country').mean().sort_values(by = ['price'],ascending = False)
top_5 = pd.DataFrame(top_5.iloc[0:5,1])
top_5['Index'] = list(range(5))
top_5['Country'] = top_5.index.values
top_5.set_index(['Index'],inplace = True)
top_5
```

```python
top_5 = train.groupby('country').mean().sort_values(by = ['price'],ascending = True)
top_5 = pd.DataFrame(top_5.iloc[0:5,1])
top_5['Index'] = list(range(5))
top_5['Country'] = top_5.index.values
top_5.set_index(['Index'],inplace = True)
top_5
```

```python
res = pd.DataFrame(train[train['variety'] == 'Pinot Noir'].groupby('points').count()).iloc[:,0]
variety_points_dist = res
res = pd.DataFrame(train[train['variety'] == 'Chardonnay'].groupby('points').count()).iloc[:,0]
variety_points_dist = pd.concat([variety_points_dist,res],axis = 1)
res = pd.DataFrame(train[train['variety'] == 'Cabernet Sauvignon'].groupby('points').count()).iloc[:,0]
variety_points_dist = pd.concat([variety_points_dist,res],axis = 1)
res = pd.DataFrame(train[train['variety'] == 'Red Blend'].groupby('points').count()).iloc[:,0]
variety_points_dist = pd.concat([variety_points_dist,res],axis = 1)
res = pd.DataFrame(train[train['variety'] == 'Bordeaux-style Red Blend'].groupby('points').count()).iloc[:,0]
variety_points_dist = pd.concat([variety_points_dist,res],axis = 1)
variety_points_dist.columns = ['Pinot Noir','Chardonnay','Cabernet Sauvignon','Red Blend','Bordeaux-style Red Blend']
variety_points_dist
```

```python
countries_list = pd.DataFrame(train.country.value_counts(dropna = True)).iloc[0:10]
countries_list['Index'] = list(range(0,10))
countries_list['Country_Name'] = countries_list.index.values
countries_list.set_index('Index',inplace = True)
countries_list.columns = ['Number Of Wines Produced','Country']

countries_list
```
